### Utiliza Python y Numpy para realizar las siguientes operaciones:
1. Dado el arreglo `a = np.array([1, 2, 3, 4])`, determina si ninguno de los elementos es cero.
2. Dado el arreglo `a = np.array([0, 0, 2, 0])`, determina si contiene algún elemento que no sea cero.
3. Analiza el arreglo `a = np.array([1, 2, np.nan, np.inf])` y determina si contiene valores que sean infinitos o valores no numéricos (NaN).
4. Dado el arreglo `a = np.array([1, 2, np.inf])`, determina si contiene valores infinitos.
5. Dado el arreglo `a = np.array([1, np.nan, 3])`, determina si contiene valores NaN.
6. Dados los arreglos `a = np.array([1, 3, 5])` y `b = np.array([2, 3, 4])`, realiza una comparación elemento a elemento contemplando mayor, mayor o igual, menor y menor o igual.
7. Crea un arreglo que contenga los valores 1, 7, 13 y 105, y determina la cantidad de memoria que ocupa el arreglo.
8. Crea un arreglo de 10 ceros; uno de 10 unos; y uno de 10 cincos.
9. Crea un arreglo que contenga todos los números enteros entre 30 y 70.
10. Crea un arreglo que contenga los números pares entre 30 y 70.
11. Crea una matriz identidad de 3x3.
12. Genera un número aleatorio entre 0 y 1.
13. Crea un arreglo de 15 números aleatorios tomados de una distribución normal estándar.
14. Crea un arreglo con valores entre 15 y 55, e imprime todos los valores excepto el primero y el último.
15. Crea un vector (arreglo) de longitud 10 con valores equitativamente distribuidos entre 5 y 50.
16. Crea un vector con valores de 0 a 20, y cambia el signo de los números de 9 a 15.
17. Convierte el arreglo `a = np.array([1, 2, 3])` a una lista de Python.
18. Convierte la lista `l = [1, 2, 3]` a un arreglo de Numpy.

20. Dado el arreglo `m = np.array([[1, 2, 3], [4, 5, 6]])`, encuentra el número de filas y columnas de la matriz.

```python
import numpy as np
```

```python
# ejercicio 1
a = np.array([1, 2, 3, 4])
#Se debe determinar si algún elemento del array es igual a cero
igual_cero = np.any(a == 0)
if igual_cero == True:
    igual_cero = "Sí"
else:    igual_cero = "No"

print(f"Existe algún elemento igual a cero: {igual_cero}")
```

```python
# ejercicio 2
a = np.array([0, 0, 2, 0])
#Se debe determinar si existe algún elemento del array que sea distinto de cero
distinto_cero = np.any(a != 0)
if distinto_cero == True:
    distinto_cero = "Sí"
else:    distinto_cero = "No"
print(f"Existe algún elemento distinto de cero: {distinto_cero}")
```

```python
# ejercicio 3
a = np.array([1, 2, np.nan, np.inf])
#Revisar si en el array existe algún elemento que sea NaN o infinito
nan_infinito = np.any(np.isnan(a)) or np.any(np.isinf(a))
if nan_infinito == True:
    nan_infinito = "Sí"
else:    nan_infinito = "No"
print(f"Existe algún elemento que sea NaN o infinito: {nan_infinito}")
```

```python
# ejercicio 4
a = np.array([1, 2, 3, np.inf])
# Revisar si en el array existe algún elemento que sea infinito
infinito = np.any(np.isinf(a))
if infinito == True:
    infinito = "Sí"
else:
    infinito = "No"
print(f"Existe algún elemento que sea infinito: {infinito}")
```

```python
# ejercicio 5
a = np.array([1, np.nan, 3])
# Revisar si en el array existe algún elemento que sea NaN
nan = np.any(np.isnan(a))
if nan == True:
    nan = "Sí"
else:    nan = "No"
print(f"Existe algún elemento que sea NaN: {nan}")
```

```python
# ejercicio 6
a= np.array([1, 3, 5])
b = np.array([2, 3, 4])
#Realizar una comparación elemento a elemento entre los arrays a y b, y determinar si son iguales, mayor que o menor que, colocar los números que se están comparando en el output
comparacion = np.where(a == b, "Iguales", np.where(a > b, "Mayor que", "Menor que"))
for i in range(len(a)):
    print(f"Comparación entre {a[i]} y {b[i]}: a es {comparacion[i]} que b")
```

```python
# ejercicio 7
#Crea un arreglo que contenga los valores: 1, 7, 13 y 10, determina el espacio en memoria que ocupa el arreglo
arreglo = np.array([1, 7, 13, 10])
espacio_memoria = arreglo.nbytes
print(f"El espacio en memoria que ocupa el arreglo es: {espacio_memoria} bytes")
```

```python
# ejercicio 8
#crea un arreglo de 10 ceros, uno de 10 unos y otro de 10 cincos
arreglo_ceros = np.zeros(10)
arreglo_unos = np.ones(10)
arreglo_cincos = np.full(10, 5)
print("Arreglo de 10 ceros:", arreglo_ceros)
print("Arreglo de 10 unos:", arreglo_unos)
print("Arreglo de 10 cincos:", arreglo_cincos)
```

```python
# ejercicio 9
# crea un arreglo con todos los números enteros del 30 al 70
arreglo_enteros = np.arange(30, 71)
print(f"El arreglo con todos los números enteros del 30 al 70 es: {arreglo_enteros}")
```

```python
# ejercicio 10
# Crear un arreglo de los número pares del 30 al 70
arreglo_pares = np.arange(30, 71, 2)
print(f"El arreglo con todos los números pares del 30 al 70 es: {arreglo_pares}")
```

```python
# ejercicio 11
# Crear una matriz de identidad 3x3
matriz_identidad = np.eye(3)
print(f"La matriz de identidad 3x3 es:\n{matriz_identidad}")
```

```python
# ejercicio 12
# Generar un número aleatorio entre 0 y 1
numero_aleatorio = np.random.rand()
print(f"El número aleatorio generado entre 0 y 1 es: {numero_aleatorio}")
```

```python
# ejercicio 13
#Crear un arreglo de 15 números aleatorios de una distribución normal estándar
arreglo_normal = np.random.randn(15)
print(f"El arreglo de 15 números aleatorios de una distribución normal estándar es:\n{arreglo_normal}")
```

```python
# ejercicio 14
#Crear un arreglo con valores entre 15 y 55 e imprime los valores menos el primero y el último
arreglo_valores = np.arange(15, 56)
print(f"El arreglo con valores entre 15 y 55 es:\n{arreglo_valores}")
print(f"\nEl arreglo sin el primer y último valor es:\n{arreglo_valores[1:-1]}")
```

```python
# ejercicio 15
# Crear un vector de longitud 10 con valores equitativamente entre 5 y 50
vector_equitativo = np.linspace(5, 50, 10)
print(f"El vector es:\n{vector_equitativo}")
```

```python
# ejercicio 16
# Crea un vector con valores del 0 al 20 y cambia de signo del 9 al 15
vector = np.arange(0, 21)
vector[9:16] = -vector[9:16]
print(f"El vector es: \n{vector}")
```

```python
# ejercicio 17
# Convierte el arreglo a = np.array([1, 2, 3, 4, 5]) a una lista de python
a = np.array([1, 2, 3, 4, 5])
lista_python = a.tolist()
print(f"La lista ahora es: {lista_python}")
```

```python
#ejercicio 18
# Dada la matriz a = np.array([[1, 2], [3, 4]]), verifica que números son pares y almacenalas en otra matriz
a = np.array([[1, 2], [3, 4]])
pares = a[a % 2 == 0]
print(f"Matriz con los números pares:\n{pares}")
```

```python
# ejercicio 19
#Convierte la lista l = [1, 2, 3] a un arreglo de numpy
l = [1, 2, 3]
print(f"La lista original es: {l}")
arreglo_numpy = np.array(l)
print(f"El arreglo de numpy es: {arreglo_numpy}")
```

```python
# ejercicio 20
#Dado el arreglo m = np.array([[1, 2, 3], [4, 5, 6]]), Encuentra el número de filas y columnas de la matriz
m = np.array([[1, 2, 3], [4, 5, 6]])
filas, columnas = m.shape
print(f"El número de filas es: {filas}")
print(f"El número de columnas es: {columnas}")
```
